### LINEAR REGRESSION

With the help of this we can  find the relation between two variables

x=independent variable

y=dependent variable

y=mx+c

y=b0+b1x this is regression line equation

b0 is the y intercept

b1 - slope of line i.e m

b0 -intercept of y i.e c

x - independant variable

```python
#define a variable and shaping it to 
x=np.array([1,2,4,5,6]).reshape(-1,1)
y=np.array([2,3,4,5,6])
```

```python
#import linear regression'
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
#prediction of model
ypred=model.predict(x)
ypred
```

X IS ALWAYS IN 2 DIMENSION i.e 2D

Y IS ALWAYS IN 1 DIMENSION i.e 1D

```python
#getting values of y intercept and slope
y_intercept=model.intercept_
print(y_intercept)

#slope
slope=model.coef_
print(slope)
```

```python
#printing prediction of random value
new_predict=model.predict([[4.2]])
print(new_predict)
```

```python
#plotting values of x and y
plt.scatter(x,y)
plt.plot(x,ypred,color="g")
plt.show()
```

1)RESIDUAL

THE DIFFERENCE BETWEEN PREDICTED VALUE AND ACTUAL VALUE IS CALLED AS RESIDUAL

2)R-SQUARE

-sum of square of residual values

-square values should be high for good fit

-formula- distance from predicted to mean / distance actual to mean

-the value of r square is always in between 0 and 1


```python
# R-SQUARE
print("R square is:",model.score(x,y))
```

### OUTLIER TREATMENT

q1 - first quartile

q3 -  last quartile

IQR - inter quartile range

IQR= Q3-Q1
      
range = q1-1.5*iqr to q3+1.5*iqr

```python
Q1=df.quantile(0.25)
print(Q1)
```

```python
Q2=df.quantile(0.50)
print(Q2)
```

```python
Q3=df.quantile(0.75)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

### FALSE-OUTLIER ABSENT , TRUE-OUTLIER TRUE

```python
(df<Q1-IQR*1.5)|(df>Q1-IQR*1.5)
```

```python
#removing the outliers
new_df=df[~((df<Q1-IQR*1.5) | (df>Q3+IQR*1.5)).any(axis=1)]
```
